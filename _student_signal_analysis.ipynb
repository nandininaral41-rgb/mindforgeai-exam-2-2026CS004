## Q3. Create a function that calculates final readiness score.

Weightage used:

Attendance = 30%

Diary entries = 10%

Tasks completed = 15%

Assignment score = 25%

Lab score = 20%

The function returns:
- Readiness score
- Performance band

```python
def readiness_score(attendance, diary, task, assignment, lab):

    score = (
        attendance * 0.30 +
        diary * 2 +
        task * 3 +
        assignment * 0.25 +
        lab * 0.20
    )

    if score >= 85:
        band = "Excellent"
    elif score >= 70:
        band = "Strong"
    elif score >= 50:
        band = "Developing"
    else:
        band = "Support"
    return score, band
```

## Q4. Use loops or apply logic to classify each student.

A loop is used to process every student record and assign:
- Excellent
- Strong
- Developing
- Support

```python
def readiness_score(attendance, diary, task, assignment, lab):

    score = (
        attendance*0.25 +
        (diary/15*100)*0.15 +
        (task/10*100)*0.20 +
        assignment*0.20 +
        lab*0.20
    )

    return round(score,2)


df["readiness_score"] = df.apply(
    lambda x: readiness_score(
        x["attendance_pct"],
        x["diary_entries"],
        x["tasks_completed"],
        x["assignment_score"],
        x["lab_score"]
    ),
    axis=1
)


df
```
